Consider a stock that has a bid-ask spread of $s = |ùëè ‚àí ùëé| > 0$. For each individual transaction that

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
def Tstar(a, b, pl, pu, pm, sigma, p0):
    tstar = ((a-b)**2 * ((pl+pu)-(pl-pu)**2)) / (4 * p0**2 * sigma**2)
    return tstar

def plot_Tstar(a, b, pl, pu, pm, sigma, p0):
    tstar = Tstar(a, b, pl, pu, pm, sigma, p0)
    plt.plot(pl, tstar, label = 'pm = ' + str(pm))
    plt.xlabel('pl')
    plt.ylabel('Tstar')
    plt.title('Tstar vs pl')
    plt.legend()
    
a = 62.05
b = 62.00
sigma = 0.2
p0 = 62.025
pm = [0, 0.1, 0.2]
```

```python
#Run a for loop to plot Tstar for each pm
pl_range = 1.1
for i in pm:
    pl = np.arange(0, pl_range, 0.1)
    plot_Tstar(a, b, pl, 1-pl-i, i, sigma, p0)
    pl_range -= 0.1
plt.show()
```
